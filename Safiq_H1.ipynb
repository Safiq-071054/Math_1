```python
while True:
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        print("Select operation:")
        print("A. Addition (+)")
        print("S. Subtraction (-)")
        print("M. Multiplication (*)")
        print("D. Division (/)")

        choice = input("Enter choice (A/S/M/D): ").upper()

        if choice == 'A':
            result = num1 + num2
            print("Result:", result)

        elif choice == 'S':
            result = num1 - num2
            print("Result:", result)

        elif choice == 'M':
            result = num1 * num2
            print("Result:", result)

        elif choice == 'D':
            if num2 != 0:
                result = num1 / num2
                print("Result:", result)
            else:
                print("Error: Division by zero is not allowed.")

        else:
            print("Invalid choice! Please select a valid operation.")

    except ValueError:
        print("Error: Please enter valid numeric values (not letters).")

    again = input("Do you want to calculate again? (Y/N): ").upper()
    if again != 'Y':
        print("Calculator exited.")
        break
```

```python

char = input("Enter a single character: ").lower()
if len(char) == 1 and char.isalpha():
    if (char == 'a' or char == 'e' or char == 'i' or char == 'o' or char == 'u'):
        print("It is a vowel.")
    else:
        print("It is a consonant.")
else:
    print("Invalid input! Please enter a single alphabet character.")
```

```python
# Taking temperature input
temp = float(input("Enter temperature: "))

print("Select conversion type:")
print("C. Celsius to Fahrenheit")
print("F. Fahrenheit to Celsius")

choice = input("Enter choice (C/F): ").upper()

if choice == 'C':
    # Celsius to Fahrenheit
    result = (temp * 9/5) + 32
    print("Temperature in Fahrenheit:", round(result, 2))

elif choice == 'F':
    # Fahrenheit to Celsius
    result = (temp - 32) * 5/9
    print("Temperature in Celsius:", round(result, 2))

else:
    print("Invalid choice! Please select C or F.")
```

```python
# Taking input from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# Finding the largest number
if num1 >= num2 and num1 >= num3:
    largest = num1
elif num2 >= num1 and num2 >= num3:
    largest = num2
else:
    largest = num3

# Display result
print("The largest number is:", largest)
```
